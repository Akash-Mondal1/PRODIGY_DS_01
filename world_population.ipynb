```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('/home/akash1/Desktop/internship/world pupulation/world_population_by_country_2023.csv')
```

```python
data.head()
data.tail()
data.describe()
data.info()
data.isnull().sum()
#Handling missing values
data['population_urban'].fillna(data['population_urban'].mean(),inplace=True)
data.isnull().sum()
data.nunique()
```

```python
# Sorting DataFrame by median_age and selecting top 15 countries
top_15_countries = data.sort_values(by='median_age', ascending=False).head(15)

# Creating bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_15_countries['country'], top_15_countries['median_age'], color='skyblue')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Median Age')
plt.title('Median Age of Top 15 Countries')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
most_dense = data[['density', 'country']].sort_values(by="density",ascending=False)
most_dense = most_dense.head(10)
most_dense = most_dense.sort_values(by="density",ascending=True)
most_dense
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='density',x='country', color="darkcyan",data=most_dense)
graph.set_title('Countries with the highest density')
plt.xticks(rotation=85)
plt.show()
```

```python
high_world_share = data[['world_share', 'country']].sort_values(by="world_share",ascending=False)
high_world_share = high_world_share.head(10)
high_world_share = high_world_share.sort_values(by="world_share",ascending=True)
high_world_share
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='world_share',x='country', color="green",data=high_world_share)
graph.set_title('Countries with the highest world share of population')
plt.xticks(rotation=85)
plt.show()
```

```python
most_pop = data.groupby(['country']).sum()['population'].reset_index()
most_pop.sort_values(by=['population'], ascending=False, inplace=True)
most_pop = most_pop.head(10)
most_pop.sort_values(by=['population'], ascending=True, inplace=True)
most_pop
```

```python
# Sorting DataFrame by median_age and selecting top 15 countries
top_15_countries = data.sort_values(by='median_age', ascending=False).tail(15)

# Creating bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_15_countries['country'], top_15_countries['median_age'], color='skyblue')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Median Age')
plt.title('Median Age of Last 15 Countries')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```
